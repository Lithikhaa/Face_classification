```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib as mpl
import os
import cv2
```

```python
import tensorflow as tf
import tensorflow.keras as keras
```

```python
import os
data_path = 'D:/face mask classification/Face-Mask-Detection-master/dataset/Models/data1'#withmask,without mask
categories = os.listdir(data_path)#dataset in list
labels = [i for i in range(len(categories))]#for eg--there are 5 categories, so len(categories) would be 5., for 5 categories, range(len(categories)) gives us [0, 1, 2, 3, 4]., i--[0, 1, 2, 3, 4]
label_dict = dict(zip(categories,labels))#zip()function to combine #dict-->eg:india;delhi(key value pair)
```

```python
len(categories)
```

```python
data = [] #imgdata and labels will be stored
target = []
for category in categories:
  #label = 0 if category == "with_mask" else 1 #loop iterates through each category in the categories list.
    folder_path = os.path.join(data_path,category)#combines the base data_path with the specific category
    img_names = os.listdir(folder_path)
    for img_name in img_names:
        img_path = os.path.join(folder_path,img_name)#creates the full path to the current image by combining folder_path and img_name.
        img = cv2.imread(img_path)
        if(img is not None):
            try: #errors may occur
                gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)#converts orginal bgr color to grayscale
                resized = cv2.resize(gray,(100,100))
                data.append(resized)#resized image is added to the data list.
                target.append(label_dict[category])
            except Expection as e:
                print('Expection:',e)
```

```python
target
```

```python
import numpy as np

data=np.array(data)/255.0
data=np.reshape(data,(data.shape[0],100,100,1))
target=np.array(target)

from keras.utils import np_utils

new_target=np_utils.to_categorical(target)
```

```python
new_target
```

```python
np.save('data',data)
np.save('target',new_target)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import optimizers
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',input_shape=data.shape[1:]))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(learning_rate=1e-4),
              metrics=['acc'])
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.33, random_state=42)
```

```python
data = np.array(data)
target = np.array(target)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

checkpoint = ModelCheckpoint(
  'model-{epoch:03d}.model',
  monitor='val_loss',
  verbose=0,
  save_best_only=True,
  mode='auto')

history=model.fit(
  train_data,
  train_target,
  epochs=15,
  callbacks=[checkpoint],
  validation_split=0.2)
```

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.show()



figure = plt.gcf() 
figure.set_size_inches(8, 6) 
#plt.savefig('filename.png', bbox_inches='tight')
plt.savefig('loss_accuracy_graph.png',dpi=300, bbox_inches='tight')
```
